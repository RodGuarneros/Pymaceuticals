### ✏️ Exploring datasets and cleaning process:
    
 Having in mind the objective: "Compare the performance of Pymaceutical's drugs of interest", based on an study with 249 mice which were monitored so as to know their evolution over the course of 45 days.
 
 The dataset "mouse_metadata" is perfect because there is not missing values, there are 249 unique mice and is ready to joint with the other dadaset.
 
 The dataset "study_results" has also valuable information regarding the evolution of each mice inside the mouse_metadata dataset, it brings us 1,893 observations relted with the evolution about every mice, but there is some risks (v.gr. duplicated observations).
 
 First, those dataset were explored with descriptive statistics before a cleaning process, following the next steps: 
 
 * Look for dimension of each dataset
 * Know, as far as possible, each dataset and the information available in each column as well as the class of information
 * Look for missing data
 * Look for duplicated observations
 * Merge both datasets with "left" criteria in order to get all the information
 * Identify every duplicate observation in Mouse ID and Timepoint
 * Drop duplicated observations so as to get a cleaned data frame. This point was crucial. Having in mind that duplicated()return a boolean serie denoting duplicate rows, see [documentation](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.duplicated.html).

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Note: First recommend taking the same statistics for complete dataset

mean_tumor = np.mean(clean_df["Tumor Volume (mm3)"])
median_tumor = np.median(clean_df["Tumor Volume (mm3)"])
variance_tumor = np.var(clean_df["Tumor Volume (mm3)"])
sd_tumor = np.std(clean_df["Tumor Volume (mm3)"])

summary_complete = pd.DataFrame(
                    {"Mean": mean_tumor,
                     "Median" : median_tumor,
                     "Variance" : variance_tumor,
                     "Standard Deviation" : sd_tumor
  }, index=["Tumor Volume (mm3)"]
)
type(summary_complete)
```

```python
clean_df.groupby(["Drug Regimen"]).sem()
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
tumor = clean_df.loc[:,["Drug Regimen","Tumor Volume (mm3)"]]
tumor_regimen = tumor.groupby(["Drug Regimen"]) 
tumor_summary = tumor_regimen.describe()
# variance
tumor_variance = tumor_regimen.var()
tumor_summary["Variance"] = tumor_variance.transform("Tumor Volume (mm3)")
tumor_median = tumor_regimen.median()
tumor_summary["Median"] = tumor_median.transform("Tumor Volume (mm3)")
# The standard error o the mean (SEM) describes how far a sample's mean is from the population's "true" mean.
# It is a function of sample size; as sample size increases, the standard error decreases. Having in mind that, we should get several samples
# This experimental data are summarized using the mean and stadard deviation of the sample data or the mean with the estandard error. Having in mind that
# the SEM is descriptive of the random sampling process. The standard error of the mean is a probabilistic statement about how the sample size will provide
# a better bound on estimates of the population mean, having in mind the central limit theorem.
# Because of that, the SEM is an estimation of how far the sample mean is likely to be from the population mean. The less the size of the sample is, the lesser is the SEM 
tumor_sem = tumor_regimen.sem()
tumor_summary["SEM"] = tumor_sem.transform("Tumor Volume (mm3)")
# making a sample set
tumor_summary
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the most promising treatment regimens. Calculate the IQR and quantitatively determine if there are any potential outliers. 
final_tumor = clean_df.loc[clean_df["Timepoint"]==45]
final_tumor = final_tumor.loc[:,["Drug Regimen","Mouse ID","Tumor Volume (mm3)"]]
#the four of th most promising treatments based on the information previously found based on the following graph: 
final_tumor.boxplot(column=['Tumor Volume (mm3)'], by='Drug Regimen', figsize = (18,6))
plt.title("Box plot of the final tumor volume  by all regimens of interest", fontsize=23)
plt.xlabel("Drug Regime", fontsize=15)
plt.ylabel("Tumor Volume (mm3)", fontsize=15)
plt.xticks(fontsize = 19)
plt.show()

# 1. Ramicane
# 2. Capomulin
# 3. Propriva
# 4. Infubinol

# let's look for outliers into promising regimens
promising = final_tumor.loc[(final_tumor['Drug Regimen']=="Ramicane") | (final_tumor['Drug Regimen']== "Capomulin") 
                | (final_tumor['Drug Regimen']== "Propriva")|(final_tumor['Drug Regimen']== "Infubinol")]

quartiles = promising["Tumor Volume (mm3)"].quantile([.25,.5,.75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = upperq-lowerq

print(f"The lower quartile of promising regimens is: {lowerq}")
print(f"The upper quartile of promising regimens is: {upperq}")
print(f"The interquartile range of promising regimens is: {iqr}")
print(f"The the median of promising regimens is: {quartiles[0.5]} ")

lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)
print(f"Values below {lower_bound} could be outliers.")
print(f"Values above {upper_bound} could be outliers.")

#is there outliers? the boxplot says there are not
# if we filter: 

outliers = final_tumor.loc[(final_tumor['Tumor Volume (mm3)']<= -8.05540533) | (final_tumor['Tumor Volume (mm3)']>= 102.30326962999999)]
outliers #this show ous that there is not observations below to lower bound and exceeding upper bound 

```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest

# Box plot for total promising regimen
red_square = dict(markerfacecolor='r', marker='s')
promising.boxplot(column=["Tumor Volume (mm3)"], figsize = (12,8),flierprops=red_square)
plt.title("Box plot of the final tumor volume of each mouse total four regimens of interest", fontsize=23)
plt.ylabel("Number of Mice", fontsize=15)
plt.xticks(fontsize = 19)
plt.show()

#Box plot each mause accross four regimens of interest
promising.boxplot(by="Drug Regimen", figsize= (12,8))
plt.title("Box plot of the final tumor volume of each mouse across four regimens of interest", fontsize=23)
plt.ylabel("Tumor Volume (mm3)", fontsize=15)
plt.xticks(fontsize = 19)
plt.show()



```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen
capomulin_df2 = clean_df.loc[:,["Drug Regimen","Mouse ID","Tumor Volume (mm3)", "Weight (g)"]]
capomulin_df2 = capomulin_df2.loc[(capomulin_df2["Drug Regimen"]=="Capomulin")]
capomulin_weight= capomulin_df2.groupby("Mouse ID").mean()

plt.scatter(capomulin_weight.iloc[:,1],capomulin_weight.iloc[:,0], c=capomulin_weight.iloc[:,0])
plt.title ("Mice Weight vs. Average Tumor Volume", fontsize=19)
plt.xlabel('Mouse Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
clb =plt.colorbar()
clb.set_label("Tumor Volume (mm3)")
plt.show()

```

## Correlation and Regression

```python
from scipy.stats import linregress
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
x_values = capomulin_weight.iloc[:,1]
y_values = capomulin_weight.iloc[:,0]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values, c=y_values)
plt.plot(x_values,regress_values, "r-")
plt.annotate(line_eq,(20,36),fontsize=15,color="green")
plt.title("Mouse Weight vs Average Tumor Volume", color="green")
plt.xlabel('Weight in grams of each Mice')
plt.ylabel('Tumor Volume (mm3)')
print(f"The r-squared is: {rvalue**2}")
plt.show()
```
